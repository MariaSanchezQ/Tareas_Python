```python
import matplotlib.pyplot as plt
vehiculos_vendidos=['Mazda cx 30','Renault kwid','Chevrolet Onix','Renault Stepway','Renault Duster','Renault Sandero']
Unidades=[4976,4713,3890,3705,3653,3469]
colores=['yellow','blue','red','orange','green','pink']
plt.pie(Unidades, labels=vehiculos_vendidos,colors=colores,wedgeprops={'edgecolor':'black'}, shadow=True)
plt.title('Comparativa vehículos vendidos')
```

```python
import matplotlib.pyplot as plt
Unidades=[4976,4713,3890,3705,3653,3469]
caballos_fuerza=[186,66,160,111,143,156]
precio_veh=[108,44,52.7,50,74,48.8]
plt.plot(Unidades,precio_veh,color='red',linewidth=3,label='Precio Vehículos (M)')
plt.plot(Unidades,caballos_fuerza,color='blue',linewidth=4,label='Caballos de fuerza')
plt.legend()
plt.title('Comparativa Unidades vs Caballos de fuerza \n\n  Unidades vs Precio de vehículos')
plt.xlabel('Unidades')
plt.ylabel('Caballos de fuerza \n\n Precio vehículos')
```
